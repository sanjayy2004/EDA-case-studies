```python
#Q1 which state is having the highest honey production throughout all these years? represent it using a bar graph
```

```python
high_prod_state = df.groupby('state')['production'].sum().sort_values(ascending=False).reset_index()
```

```python
high_prod_state
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=high_prod_state, x='production',y='state',palette='hsv')
plt.show()
```

```python
#Q2)which state is having the highest and the lowest yeild per colony?
#represent it using the bar graph
```

```python
highest_lowest_yeild = df.groupby(['state'])['yield_per_colony'].mean().sort_values(ascending=False).reset_index()
```

```python
highest_lowest_yeild
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=highest_lowest_yeild,x='yield_per_colony',y='state',palette='hsv')
plt.show()
```

```python
#Q3.which year wittnessed the highest in honey production? use bar graph to plot it.
```

```python
high_honey_prod_year = df.groupby('year')['value_of_production'].sum().sort_values(ascending=False).reset_index()
```

```python
high_honey_prod_year
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=high_honey_prod_year,x='year',y='value_of_production',palette='hsv')
plt.xticks(rotation=90)
plt.show()
```
