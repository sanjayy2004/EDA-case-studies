```python
#create a function which will take outliers column list and df as parameter and then it will return the outlier free data st
```

```python
out_list =['Purchase']
```

```python
def outlier_eliminator(df,out_list):
    for col in out_list:
        Q1 = df[col].quantile(0.25)
        Q2 = df[col].quantile(0.75)
    
        IQR = Q3-Q1
    
        LB = Q1-1.5*(IQR)
        UB = Q3+1.5*(IQR)
    
        df=df[(df[col]>=LB) & (df[col]<=UB)]
    return df
    
```

```python
df = outlier_eliminator(df,out_list)
df.shape
```
